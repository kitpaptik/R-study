```r
#making new function
#you can make new function by using function()

mypower = function(x){
    result = c(x^2, x^3) #x squared, x cubed
    return(result)
}
mypower(3)
```

# Conditional Functions

```r
#if... else...

mypower = function(x){
    if(is.numeric(x)){
        result = c(x^2, x^3)
    }
    else{
        result = NA
    }
    result
}
mypower(3)
```

```r
mypower("moon")
```

```r
#ifelse()

mypower2 = function(x){
    ifelse(is.numeric(x), c(x^2, x^3), NA) #if x is numeric return c(x^2, x^2), else NA
}
mypower2(3)
```

```r
mypower2("moon")
```

```r
#for

myfact = function(x){
    result = 1
    for(i in 1:x){
        result = result*i
    }
    result 
}
myfact(3)
```

```r
myfact(5)
```

```r
#you can use vector or function in for loop list

par(mfrow = c(2,2))
x = list(1:6, sin, function(x) {x^2+2*x}, dnorm)
for(i in x) plot(i, xlim = c(0, 2*pi))
par(mfrow = c(1, 1))
```

```r
#while
#loops until certain condition

i = 1
while(i < 5){
    i = i + 1
}
i
```

```r
#break
#finish the loop

i = 1
while(1){ #infinite loop
    if(i == 5) break
    i = i + 1
}
i
```

```r
#next
#skip to next rep

oddSum = 0
for(i in 1:100){
    if(i%%2 == 0) next
    oddSum = oddSum + i
}
oddSum
```

```r
#repeat()
#same as while(1)

i = 1
repeat{
    if(i == 5) break
    i = i + 1
}
i
```
