# Operations Used in R

```r
#getting area of a circle with radius length 2cm

r = 2 #radius
circle = pi * r^2 
circle
```

```r
#the difference between = and ==
#= is setting the variable with a certain value
#== is checking if a variable is same with a certain value

r == 2 #is r same with 2?
```

```r
#what happens if the variable is not same with the value?

r = 3
r == 2
```

```r
#!= means 'not'

r != 2 #r is not same with 2
```

```r
#<- operation is for setting a value for many variables

a<-b<-c<-d<-2 #put 2 for a,b,c,d
a + b + c + d
```

```r
#example of getting an average

total = 2680; n = 31 #using ; can allow writing two short lines in a single line
average <- total/n
average
```

```r
#when writing a power of a number, use ^ or **
#ex) 2 cubed is 2**3 or 2^3
#%% represents for remainder

5 ** 2 #5 square
10 %% 3 #remainder of 10 divided by 3
```

```r
#parenthesis are treated equally as it is in standard math

d = (1+2) * 3
d
```

```r
print(d)

#as shown before, if you do not write anything but the name of a variable, R automatically recognize it as print(variable)
```

```r
e = d - 7
print(e) #print(e) and e is equal
```

```r
e
```

```r
#you can shorten setting a value and printing the variable
(e = d - 7) #e = d - 7; e
```

```r
a = c(1, 2, 3, 10, 6) #combine(), or c(), can assign many different values into one variable
mode(a) #type of values in a
```

```r
str(a) #structure of a
```

```r
b = c(7, 6)
a = c(a, 0) #adding values of a and 0 to a
c = c(a, b, 7, -5) #adding values of a , values of b, 7, and -5 to c
```

```r
length(c) #number of values in c
```
