# Factors

```r
sex = c("Male", "Female", "Female", "Male", "Male")

#when a data is entered like this, vector sex is counted as string vector
```

```r
sex = factor(sex)
sex
```

```r
str(sex)
```

```r
levels(sex)
```

```r
length(sex)
```

```r
#Vector can be factored afterwards

smoking = c(1, 1, 2, 3, 1)
smoking = factor(smoking)
levels(smoking) = c("none", "ex-smoker", "smoker") #none: 1, ex-smoker: 2, smoker: 3
smoking
```

```r
Height = c(168, 173, 160, 145, 180)
Weight = c(80, 65, 92, 53, 76)
BMI = Weight/(Height/100)^2
BMI
```
