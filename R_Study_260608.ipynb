# Collecting Random Datas

```r
#if a certain data is too broad, sometimes we need to do uneccessarily long math operations. In these cases,
#using random data collection can save general characteristics of a data set, and make the operation faster.

#ex) getting a lotto number
S1 = sample(1:45, 6, replace = FALSE) #Picking 6 random numbers from 1~45
```

```r
print(S1)
```

```r
#functions like sample(), which return random number, can be pre-set by using set.seed()

set.seed(1234)
S2 = sample(1:45, 6, replace = FALSE)
```

```r
print(S2)

#any number can be inside the set.seed() function
#in this case, there is a certain numbers already assigned to the seed 1234, which would allow us to know the return value before running the code
```

# Using if() Function

```r
#sometimes we need to repeat a certain action or check the values case by case while analyzing a data set
#in these cases, rather than checking them one by one, using for() or if() function can be efficient

A = c(1,2,3,4,5) 
if(7 %in% A){ # %in% checking if a value is in A
    print("TRUE")
} else{
    print("FALSE")
}
```

# Using function() to make user-function

```r
#most of the data analyzing can be done with pre-loaded R functions
#however, sometimes we need to make our own functions

Plus_One = function(x){
    y = x+1
    return(y)
}
```

```r
Plus_One(3)
```
