# Handling a data set

```r
#ifelse() : checking one condition. if it is false, do another action

HR$satisfaction_level_group_1 = ifelse(HR$satisfaction_level > 0.5, 'High', 'Low')
#if satisfaction_level is higher than 0.5, mark it as high. if not, mark it as low
```

```r
#ifelse can be inside another ifelse : ifelse(ifelse())

HR$satisfaction_level_group_2 = ifelse(HR$satisfaction_level > 0.8, 'High',
                                   ifelse(HR$satisfaction_level > 0.5,'Mid','Low'))

HR$satisfaction_level_group_2 = as.factor(HR$satisfaction_level_group_2)
```

```r
#subset() : extracting a certain value

HR_High = subset(HR, salary == 'high')
```

```r
#Making a new data set with people who have high salary and their sales are IT

HR_High_IT = subset(HR,salary == 'high' & sales == 'IT')
```

```r
#New data set with people who have high salary or their sales are IT

HR_High_IT2 = subset(HR,salary == 'high' | sales == 'IT')
```
