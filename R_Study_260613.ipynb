# NA and NULL

```r
a = as.numeric("Kim") #if a non numeric value is transformed to an integer, the number changes to NA(not available)
```

```r
a
```

```r
#NA is not same with NULL
#NULL means that a variable is empty

a = c()
a
```

```r
a = c(1, 2, 3, NA, 8, NA, 10) #2 NAs in a
mean(a)
```

```r
mean(a, na.rm = T) #removing NA values from a
```

```r
sum(is.na(a)) #counting numbers of na in a
```

```r
a = c(a, NULL, 8, NULL);a #there is no error if there is null
```

```r
(a=c(a,NA))
```

```r
a = NULL #how to delete values in a
a
```
