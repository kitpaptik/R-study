# Group variables and row variables

```r
#On the left isde of ~, we write teh grouping variable
#On the right side, write the row variabales
#write + to add or - to remove variable name
mytable(sex~age+DX, data = acs)
```

```r
mytable(am~.-hp-disp-cyl-gear, data = mtcars)
```

# Continuous Variable Statistics

```r
#in most of the cases, if not pre-chosen, continuous variables are treated as normal distribution
#i will use another data set, radial, to make another table

data(radial)
mytable(sex~age+weight+TC+hsCRP, data = radial, method = 3)
```

# Categorical Variable / Continuous Variable

```r
#most of times, we input numbers as options
#ex) smoking: 1, non-smoking: 2

mytable(am~., data = mtcars)
```

```r
#as shown, cyl, vs, and gear variables are treated as categorical variables
#carb variable also has 6 eigenvalues, there fore to change carb to categorical variable, we need to change max.ylev

mytable(am~carb, data = mtcars, max.ylev = 6)
```

# How to Combine Tables

```r
#sometimes, you need to combine many different tables into one
#if you wanted to make two separate tables and combine into one table,
#mytable is the function of choice

mytable(sex+DM~., data = acs)
```

```r
#this is also allowed, but the tables becomes to messy

mytable(sex+DX~., data = acs)
```

# How to make High Quality Table: mylatex

```r
#Latex should be intalled on the computer to run the mylatex code
out = mytable(Dx~., data = acs);mylatex(out)
```

# Downloading file as CSV

```r
mycsv(out, file = "test1.csv")
mycsv(mytable(sex+DM~age+Dx, data = acs), file = "test2.csv")
```

# Printing in HTML type table

```r
#myhtml() function can print the table in the form of what we see in the web

out = mytable(sex~age+Dx, data = acs)
myhtml(out)
myhtml(mytable(sex+DM~age+Dx, data = acs))
```
